```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('garfico_ocorrencias.csv', low_memory = False, delimiter = ';', encoding = 'UTF-8', decimal = ',')
df
colunas = []
a = df.iloc[:, 0]
b = df.iloc[:, 1]
fig, ax = plt.subplots(figsize = (15, 8), dpi = 300)
plt.barh(a,b)
plt.xlim(0, 800)
plt.ylabel('Recorrências', fontsize=15)
plt.xlabel('Ocorrência das recorrências', fontsize=15)
plt.title('Distribuição dos chamados recorrentes', fontsize=15)
```
